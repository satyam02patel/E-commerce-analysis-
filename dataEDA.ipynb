```python


# Prepare empty list for results
ct_results = []

for col in df.columns:
    col_data = df[col]
    entry = {'Column Name': col}

    # Numeric columns
    if pd.api.types.is_numeric_dtype(col_data):
        entry['Mean'] = round(col_data.mean(), 2)
        entry['Median'] = round(col_data.median(), 2)
        entry['Mode'] = col_data.mode().iloc[0] if not col_data.mode().empty else None
        

    # Categorical columns
    else:
        entry['Mean'] = None
        entry['Median'] = None
        entry['Mode'] = col_data.mode().iloc[0] if not col_data.mode().empty else None
        

    ct_results.append(entry)

# Convert to DataFrame
ct_df = pd.DataFrame(ct_results)

ct_df
```

```python


# List of numeric columns for dispersion analysis
numeric_cols = [
    'Age', 'Price', 'Quantity', 'Discount(%)', 'Rating', 
    'ShipmentTime(days)', 'LateDelivery_encoded', 'LastLogin(days)', 
    'CartAbandonmentRate', 'ClicksPerSession', 'Revenue', 
    'ReturnReason_encoded', 'ChurnRiskScore'
]

# Create a DataFrame to store the results
dispersion_df = pd.DataFrame(columns=['Column', 'Variance', 'Standard Deviation', 'Range'])

# Calculate dispersion metrics
for col in numeric_cols:
    if col in df.columns:
        variance = df[col].var()
        std_dev = df[col].std()
        value_range = df[col].max() - df[col].min()
        dispersion_df = pd.concat([dispersion_df, pd.DataFrame({
            'Column': [col],
            'Variance': [variance],
            'Standard Deviation': [std_dev],
            'Range': [value_range]
        })], ignore_index=True)

# Display the result
print(dispersion_df)
```

#skewness

```python
cols = ['Age', 'Price', 'Quantity', 'Discount(%)', 'Rating', 
        'ShipmentTime(days)',  'LastLogin(days)', 
        'CartAbandonmentRate', 'ClicksPerSession', 'Revenue', 
        'ReturnReason_encoded', 'ChurnRiskScore']

skewness_values = {}

for col in cols:
    skewness_values[col] = df[col].skew()

# Print skewness for each column
for col, skew_val in skewness_values.items():
    print(f"{col} skewness: {skew_val}")
```

####rules
skewness < 0.5: Data roughly symmetric, generally no action needed.

0.5 â‰¤ skewness < 1: Moderate skewness, consider transformation if the model requires.

skewness â‰¥ 1: High skewness, strongly consider transformation

```python
import numpy as np

# Adding 1 to avoid log(0)
df['LateDelivery_encoded_log'] = np.log1p(df['LateDelivery_encoded'])
df['Revenue_log'] = np.log1p(df['Revenue'])

# Check skewness after transformation
print("Skewness after log transform:")
print("LateDelivery_encoded_log:", df['LateDelivery_encoded_log'].skew())
print("Revenue_log:", df['Revenue_log'].skew())
```

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')
df['LateDelivery_encoded_yeojohnson'] = pt.fit_transform(df[['LateDelivery_encoded']])

print("Skewness after Yeo-Johnson transform:", df['LateDelivery_encoded_yeojohnson'].skew())
```

```python
import pandas as pd
from scipy.stats import kurtosis

# List of numerical columns for kurtosis calculation
numerical_cols = [
    'Age', 'Price', 'Quantity', 'Discount(%)', 'Rating', 'ShipmentTime(days)',
    'PurchaseFrequency_encoded', 'CartAbandonmentRate',
    'ClicksPerSession', 'Revenue', 'ChurnRiskScore'
]

# Convert specified columns to numeric (force non-numeric to NaN)
df[numerical_cols] = df[numerical_cols].apply(pd.to_numeric, errors='coerce')

# Calculate kurtosis and type
kurtosis_values = {}
kurtosis_types = {}

for col in numerical_cols:
    col_data = df[col].dropna()
    kurt_val = kurtosis(col_data, fisher=False)  # Raw kurtosis: 3 = normal
    kurtosis_values[col] = kurt_val

    # Classify kurtosis type
    if kurt_val < 3:
        kurt_type = "Platykurtic"
    elif kurt_val > 3:
        kurt_type = "Leptokurtic"
    else:
        kurt_type = "Mesokurtic"
    
    kurtosis_types[col] = kurt_type

# Convert to DataFrame
kurtosis_df = pd.DataFrame({
    'Kurtosis': pd.Series(kurtosis_values),
    'KurtosisType': pd.Series(kurtosis_types)
})

# Show result
print(kurtosis_df)
```

```python
total_cells = df.size
missing_cells = df.isnull().sum().sum()
completeness = ((total_cells - missing_cells) / total_cells) * 100
```

```python
valid_quantity = df[df['Quantity'] > 0].shape[0]
validity = (valid_quantity / df.shape[0]) * 100
```

```python
unique_orders = df['OrderID'].nunique()
uniqueness = (unique_orders / df.shape[0]) * 100
```

```python
consistent_gender = df[(df['Gender'] == 'Male') & (df['Gender_encoded'] == 1) |
                       (df['Gender'] == 'Female') & (df['Gender_encoded'] == 0)].shape[0]
consistency = (consistent_gender / df.shape[0]) * 100
```

```python
data_quality_score = (completeness + validity + uniqueness + consistency) / 4
```

```python

# 1. Line ChartRevenue Trend Over Time (Monthly)
df['PurchaseMonth'] = df['PurchaseDate'].dt.to_period('M').astype(str)
monthly_revenue = df.groupby('PurchaseMonth')['Revenue'].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_revenue, x='PurchaseMonth', y='Revenue', marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
