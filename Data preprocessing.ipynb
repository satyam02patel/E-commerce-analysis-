```python
# List of columns to check for outliers
cols_to_check = ['Price', 'Revenue', 'ShipmentTime(days)', 'ClicksPerSession']

# Loop through each column and remove outliers using IQR
for col in cols_to_check:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Keep only rows within the IQR bounds
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

df
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder



# Optional: clean column names (strip spaces, lowercase)
df.columns = df.columns.str.strip()

# Initialize LabelEncoders for categorical columns
le_gender = LabelEncoder()
le_reason = LabelEncoder()

# 1. Encode Gender (object)
df['Gender_encoded'] = le_gender.fit_transform(df['Gender'])

# 2. Encode LateDelivery and Returned (bool to int)
df['LateDelivery_encoded'] = df['LateDelivery'].astype(int)
df['Returned_encoded'] = df['Returned'].astype(int)

# 3. Map PurchaseFrequency (ordinal)
freq_map = {'low': 0, 'medium': 1, 'high': 2}
df['PurchaseFrequency_encoded'] = df['PurchaseFrequency'].str.strip().str.lower().map(freq_map)


# 4. Handle missing values in ReturnReason before encoding
# Fill missing ReturnReason with a placeholder, e.g. 'No_Return'
df['ReturnReason'] = df['ReturnReason'].fillna('No_Return')

# Label Encode ReturnReason
df['ReturnReason_encoded'] = le_reason.fit_transform(df['ReturnReason'])

# Check the encoded columns
print(df[['Gender', 'Gender_encoded', 'LateDelivery', 'LateDelivery_encoded',
          'PurchaseFrequency', 'PurchaseFrequency_encoded',
          'Returned', 'Returned_encoded',
          'ReturnReason', 'ReturnReason_encoded']].head())
```

```python

df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 60, 100], labels=['Young', 'Middle-aged', 'Senior'])
df
df['RatingGroup'] = pd.cut(df['Rating'], bins=[0, 2, 4, 5], labels=['Low', 'Average', 'High'])
df
```

```python
# Discount binning based on actual data
df['DiscountGroup'] = pd.cut(
    df['Discount(%)'],
    bins=[0, 5, 15, 19, df['Discount(%)'].max()],
    labels=['No discount', 'Low', 'Medium', 'High'],
    include_lowest=True
)

# View output
print(df[['Discount(%)', 'DiscountGroup']].head(10))


df.head(21)
```
